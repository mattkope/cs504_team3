```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("matt2434/cleaned-ecommerce-data")

print("Path to dataset files:", path)
```

```python
import pandas as pd
import os

# Storing File Paths
nov_name = '2019-Nov-Cleaned.csv'
nov_path = os.path.join(path, nov_name)

dec_name = '2019-Dec-Cleaned.csv'
dec_path = os.path.join(path, dec_name)

# Loading File Paths
nov = pd.read_csv(nov_path)
dec = pd.read_csv(dec_path)

# Check when complete
dec.tail()
```

```python
# Combining Datasets
df = pd.concat([nov, dec], ignore_index=True)
df.tail()
```
