```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("matt2434/cleaned-ecommerce-data")

print("Path to dataset files:", path)
```

```python
import pandas as pd
import os

# Storing File Paths
nov_name = '2019-Nov-Cleaned.csv'
nov_path = os.path.join(path, nov_name)

dec_name = '2019-Dec-Cleaned.csv'
dec_path = os.path.join(path, dec_name)

# Loading File Paths
nov = pd.read_csv(nov_path)
dec = pd.read_csv(dec_path)

# Check when complete
dec.tail()
```

```python
# Combining Datasets
df = pd.concat([nov, dec], ignore_index=True)
df.tail()
```

```python
df_purchase = df[df['event_type'] == 'purchase']
df_purchase.head(15)
```

```python
# Group by session and product
basket = df_purchase.groupby(['user_session', 'product_id'])['event_type'].count().unstack().reset_index().fillna(0)

# Convert quantities into binary (0 or 1) format
basket.set_index('user_session', inplace=True)
basket = basket.applymap(lambda x: 1 if x > 0 else 0)
```
