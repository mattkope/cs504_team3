```python
import pandas as pd
import os

# Combine all CSV files in the directory into a single DataFrame
file_paths = []  # Specify the path where your data files are stored

for dirname, _, filenames in os.walk('/kaggle/input'):  # Update the path if needed
    for filename in filenames:
        file_paths.append(os.path.join(dirname, filename))

# Load all CSV files and combine them
dataframes = [pd.read_csv(file) for file in file_paths if file.endswith(".csv")]
combined_df = pd.concat(dataframes, ignore_index=True)

# Verify the combined DataFrame
print(combined_df.head())
```

```python
import matplotlib.pyplot as plt

event_counts = combined_df['event_type'].value_counts()

plt.figure(figsize=(8, 8))  # Set the figure size
plt.pie(event_counts, labels=event_counts.index, autopct='%1.1f%%', startangle=90)

plt.title('Distribution of Event Types')

plt.show()
```
